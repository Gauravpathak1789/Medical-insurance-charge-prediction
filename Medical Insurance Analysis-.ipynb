# UNIVARIATE ANALYSIS-

```python
#value counts of object features
def countplot_of_categorical_features(i):
    print("Different value counts of ",i,"column--")
    print(df[i].value_counts())
    sns.countplot(df[i])
    plt.show()
    
```

```python
cols=["sex","smoker","region","children"]
for i in cols:
    countplot_of_categorical_features(i)
    print("------------------------X--------------------------")
```

# Now we can see that in terms of percentage--

```python
def pie_chart_categorical_column(n):
    print("reperesentation of ",n,"column in terms of percentage-")
    df['sex'].value_counts().plot(kind='pie',autopct='%.2f')
    plt.show()
    
    
    
```

```python
cols=["sex","smoker","region"]
for i in cols:
    pie_chart_categorical_column(i)
    print("------------------------X--------------------------")
```
